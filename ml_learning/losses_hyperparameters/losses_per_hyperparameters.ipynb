```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 1 #note: 0 is the training loss and 1 is the validation loss
```

```python
baseline = np.loadtxt("losses_baseline.txt")[i]
weightm1 = np.loadtxt("losses_-1weight.txt")[i]
weight1 = np.loadtxt("losses_1weight.txt")[i]
lr2 = np.loadtxt("losses_1e-2lr.txt")[i]
lr4 = np.loadtxt("losses_1e-4lr.txt")[i]
epoch200 = np.loadtxt("losses_200epoch.txt")[i]
epoch300 = np.loadtxt("losses_300epoch.txt")[i]
batch228 = np.loadtxt("losses_228batch.txt")[i]
batch284 = np.loadtxt("losses_284batch.txt")[i]
```

```python
x = list(range(0,len(baseline)))
```

```python
plt.plot(x, baseline, label = "Batch size = 256", color = "black")
plt.plot(x, batch228, label = "Batch size = 228")
plt.plot(x, batch284, label = "Batch size = 284")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend()
plt.yscale("log")
plt.title("Effect of Batch Size")
plt.show()
```

```python
plt.plot(x, baseline, label = "Number of epochs = 250", color = "black")
plt.plot(list(range(0,len(epoch200))), epoch200, label = "Number of epochs = 200")
plt.plot(list(range(0,len(epoch300))), epoch300, label = "Number of epochs = 300")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend()
plt.yscale("log")
plt.title("Effect of Number of Epochs")
plt.show()
```

```python
plt.plot(x, baseline, label = "Learning Rate = 1e-3", color = "black")
plt.plot(x, lr2, label = "Learning Rate = 1e-2")
plt.plot(x, lr4, label = "Learning Rate = 1e-4")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend()
plt.yscale("log")
plt.title("Effect of Learning Rate")
plt.show()
```

```python
plt.plot(x, baseline, label = "Weight decay = 0", color = "black")
plt.plot(x, weightm1, label = "Weight decay = -1")
plt.plot(x, weight1, label = "Weight decay = 1")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend()
plt.yscale("log")
plt.title("Effect of Weight Decay")
plt.show()
```
