```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 1 #note: 0 is the training loss and 1 is the validation loss
```

```python
baseline = np.loadtxt("losses_15patience.txt")[0]
x = list(range(0,len(baseline)))
```

```python
def get_data(filename):
    return np.loadtxt(filename)[i]
```

```python
r = [5,10,15,20,25]
lr_3 = ["losses_"+str(n)+"patience.txt" for n in r]
lr_2 = ["losses_"+str(n)+"patience_2lr.txt" for n in r]
lr_4 = ["losses_"+str(n)+"patience_4lr.txt" for n in r]
```

```python
def plot_1lr(lr):
    if lr == 3: 
        d = lr_3
    elif lr == 2: 
        d = lr_2
    elif lr == 4: 
        d = lr_4
    else:
        return 
    for n in range(len(r)):
        plt.plot(x, get_data(d[n]), label = "Patience = "+ str(5* (n+1)), alpha = 0.75)
        plt.yscale("log")
    plt.title("lr = 1e-" + str(lr) )
    plt.legend()
    plt.xlabel("Epoch")
    if i == 0:
        plt.ylabel("Training Loss")
    if i == 1: 
        plt.ylabel("Validation Loss")
```

```python
plot_1lr(3)
```

```python
plot_1lr(2)
```

```python
plot_1lr(4)
```
