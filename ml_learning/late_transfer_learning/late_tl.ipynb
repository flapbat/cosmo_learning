```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 0 #note: 0 is the training loss and 1 is the validation loss
```

```python
def plot_graph(tester):
    baseline_losses = np.loadtxt(tester+"_none.txt")[i]
    late1_losses = np.loadtxt(tester+"_late1.txt")[i]
    late2_losses = np.loadtxt(tester+"_late2.txt")[i]
    late3_losses = np.loadtxt(tester+"_late3.txt")[i]
    late4_losses = np.loadtxt(tester+"_late4.txt")[i]
    final = np.loadtxt(tester+"_input_output.txt")[i]
    x = list(range(0,len(baseline_losses)))
    #linestyles : https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
    plt.plot(x, baseline_losses, label = "Baseline", color = "black")
    plt.plot(x, late1_losses, label = "Late 1 (model.4 frozen)", linestyle='dotted', color = "blue")
    plt.plot(x, late2_losses, label = "Late 2 (model.3,4)", linestyle='dashed', color = "green")
    plt.plot(x, late3_losses, label = "Late 3 (model.2,3,4)", linestyle='dashdot', color = "red")
    plt.plot(x, late4_losses, label = "Late 4 (model.1,2,3,4)", dashes=[1, 5], color = "purple")
    plt.plot(x, final, label = "Input + Output (model.0,4)", dashes=[1, 1], color = "orange")

    
    plt.xlabel("Epoch")
    plt.ylabel("Training Loss")
    plt.legend()
    plt.yscale("log")
    plt.show()
```

```python
plot_graph("testing_metrics")
```
